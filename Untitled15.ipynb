# № 1.Как связаны главные компоненты с исходными данными
Главные компоненты (PCA) связаны с исходными данными через линейные преобразования. PCA представляет собой статистический метод, который используется для уменьшения размерности данных, сохраняя при этом максимальное количество информации. Он делает это, создавая новые координатные оси (главные компоненты), которые являются линейными комбинациями исходных признаков, и которые объясняют максимальное количество дисперсии в данных. Каждая главная компонента ортогональна (перпендикулярна) остальным компонентам, и они упорядочены по убыванию объясняемой дисперсии.


# № 2.
Для оценки сжатия данных с использованием метода главных компонент (PCA) в Python, мы можем рассчитать отношение объема исходных данных к объему данных после применения PCA. Исходный объем данных равен произведению количества строк на количество столбцов матрицы. После применения PCA, объем данных равен сумме количества строк, умноженного на количество главных компонент, и количества главных компонент, умноженного на количество столбцов исходной матрицы (для хранения самих компонент и коэффициентов).


```python
# Исходные размеры матрицы
rows = 4250
columns = 7

# Размеры после PCA
components = 3

# Исходный объем данных
original_volume = rows * columns

# Объем данных после PCA
pca_volume = (rows * components) + (components * columns)

# Оценка сжатия
compression_ratio = pca_volume / original_volume

print(f"Исходный объем данных: {original_volume}")
print(f"Объем данных после PCA: {pca_volume}")
print(f"Сжатие: {compression_ratio:.3f}")
```

# №3 Для генерации данных в виде эллипса с заданными параметрами можно использовать следующий код на Python

```python
import numpy as np

# Количество точек
num_points = 1100
# Центр эллипса
center = np.array([1.5, -2.5])
# Радиусы по осям x и y
radii = np.array([3, 2.5])
# Угол поворота эллипса в радианах
angle = 65 * np.pi / 180

# Матрица поворота
rotation_matrix = np.array([[np.cos(angle), -np.sin(angle)],
                            [np.sin(angle), np.cos(angle)]])

# Генерация точек вокруг центра
theta = np.linspace(0, 2 * np.pi, num_points)
x = radii[0] * np.cos(theta)
y = radii[1] * np.sin(theta)
points = np.vstack((x, y)).T

# Применение поворота и сдвига к точкам
rotated_points = points.dot(rotation_matrix) + center

# Вывод результатов
for point in rotated_points:
    print(point)
```
