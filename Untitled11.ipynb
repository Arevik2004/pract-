# №7.Чтобы найти частные производные сложной функции E по переменным w и b, мы можем использовать цепное правило дифференцирования.
Пусть y' = w*x + b.
Тогда E = (y' - y)^2.

Для нахождения частной производной функции E по переменной w, мы должны сперва найти частную производную функции E по переменной y', а затем умножить эту производную на частную производную переменной y' по переменной w.

∂E/∂w = ∂E/∂y' * ∂y'/∂w

Найдем частную производную функции E по переменной y':

∂E/∂y' = 2(y' - y)

Теперь найдем частную производную переменной y' по переменной w:

∂y'/∂w = x

Итак, подставляя найденные значения, получаем:

∂E/∂w = 2(y' - y) * x

Аналогично, чтобы найти частную производную функции E по переменной b, мы можем использовать тот же метод:

∂E/∂b = ∂E/∂y' * ∂y'/∂b

Найдем частную производную функции E по переменной y':

∂E/∂y' = 2(y' - y)

Теперь найдем частную производную переменной y' по переменной b:

∂y'/∂b = 1

Итак, подставляя найденные значения, получаем:

∂E/∂b = 2(y' - y) * 1





```python
# №8 например  решаем задачу через питон
def partial_E_w(w, b, x, y):
    return 2*(w*x + b - y)*x

def partial_E_b(w, b, x, y):
    return 2*(w*x + b - y)

w = 2
b = 3
x = 5
y = 7

print("∂E/∂w =", partial_E_w(w, b, x, y))
print("∂E/∂b =", partial_E_b(w, b, x, y))
```

# №9 Найдите частные производные сложной функции
E = y*ln(y') + (1-y)*ln(1-y'), y'=1/(1+e^(-z))
по переменной z. e - это основание натурального логарифма, y' - это предсказание модели y

Для нахождения частной производной функции E по переменной z, сначала найдем производную функции E по y' и затем найдем производную y' по z.

Найдем частную производную функции E по y':

dE/dy' = ln(y') + y/y' - ln(1-y') - (1-y)/(1-y')

dE/dy' = ln(y') + y/y' - ln(1-y') + (y-1)/(1-y')

Теперь найдем производную y' по z:

y' = 1/(1+e^(-z))

dy'/dz = 0 - e^(-z)/(1+e^(-z))^2

dy'/dz = - e^(-z)/(1+e^(-z))^2

Наконец, найдем частную производную функции E по переменной z с использованием правила дифференцирования сложной функции:

dE/dz = dE/dy' * dy'/dz

dE/dz = (ln(y') + y/y' - ln(1-y') + (y-1)/(1-y')) * (- e^(-z)/(1+e^(-z))^2)

Таким образом, мы нашли частную производную сложной функции E по переменной z.

```python
import math

# Функция вычисления уровня ошибки E
def E(y, z):
    y_predict = 1 / (1 + math.exp(-z))
    E = y*math.log(y_predict) + (1-y)*math.log(1-y_predict)
    return E

# Находим частную производную E по переменной z
def partial_E_z(y, z):
    y_predict = 1 / (1 + math.exp(-z))
    partial_E_z = y / (1 + math.exp(-z)) - (1-y) * math.exp(-z) / (1 + math.exp(-z))
    return partial_E_z

# Заданные значения
y = 0.7
z = 1.5

# Вывод результата
print("E =", E(y, z))
print("∂E/∂z =", partial_E_z(y, z))
```
